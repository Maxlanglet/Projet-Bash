### Pouvons nous justifier ces conditions aux bords ? 

La "no-slip" condition implique que l'on peut délimiter une fine couche le long des bords de notre domaine à l'intérieur de laquelle la vitesse relative de notre fluide par rapport aux parois est nulle. Etant donné que nous nous plaçons dans un cadre de haute viscosité (petit nombre de Reynolds), cette hypothèse semble justifiée. 

On a donc que $\overrightarrow{u}=(u_x,u_y,u_z)=0$ sur les parois latérales et inférieurs de notre domaine et que $u_y=1$ le long de la paroi supérieur.

Or étant donné que nous nous plaçons dans un cadre à deux dimensions, on sait que

\begin{equation}
u_x=\frac{\partial\psi}{\partial y}
\end{equation}

\begin{equation}
u_y=-\frac{\partial\psi}{\partial x}
\end{equation}

ce qui nous fournit directement nos conditions aux bords de Neumann et nous assure que chacuns des bords de notre domaine soient des equipotentiels ce qui nous fournit donc aussi nos conditions aux bords de Dirichlet vu que le choix de notre constante de départ est arbitraire.

### Différence centrale finie discrète d'ordre 2 pour l'équation biharmonique. 

Nous avons, en dévellopant le double Laplacien, l'équation suivante :

\begin{equation}
\nabla^4 \psi= \frac{\partial^4 \psi}{\partial x^4}+\frac{\partial^4 \psi}{\partial y^4}+2\frac{\partial^4 \psi}{\partial x^2 \partial y^2}
\end{equation}

De discrétiser ceci nous a donner plus de difficulters que prévus, nous avons alors préférer un différente approche pour résoudre ce problème. Voyons le Laplacien comme un opérateur étant appliqué 2 fois sur notre fonction. En discrétisant l'opérateur, on trouve alors comme expression :

\begin{equation}
\nabla^2 \nabla^2 \psi = \frac{\nabla^2\psi_{i-1,j}+\nabla^2\psi_{i+1,j}-2\nabla^2\psi_{i,j}}{\Delta x^2}+\frac{\nabla^2\psi_{i,j-1}\nabla^2\psi_{i,j+1}-2\nabla^2\psi_{i,j}}{\Delta y^2}
\end{equation}

En prenant pour des écarts égaux $\Delta x=\Delta y$:

\begin{equation}
\nabla^2 \nabla^2 \psi = \frac{\nabla^2\psi_{i-1,j}+\nabla^2\psi_{i+1,j}+\nabla^2\psi_{i,j-1}\nabla^2\psi_{i,j+1}-4\nabla^2\psi_{i,j}}{\Delta x^2}
\end{equation}

On a ainsi discrétisé le premier Laplacien, il faut ensuite faire de même pour le second. On trouve au final par le même procédé l'équation suivante : 

\begin{equation}
\nabla^4 \psi = \frac{\psi_{i-2,j}-8\psi_{i-1,j}+20\psi_{i,j}-8\psi_{i+1,j}+\psi_{i+2,j}+\psi_{i,j-2}-8\psi_{i,j-1}-8\psi_{i,j+1}+\psi_{i,j+2}}{\Delta x^2}
\end{equation}

Avec les conditions aux bords nous pourrons compléter les éléments du bords de notre opérateur. En se référant à la figure de plus tôt, nous tirons les informations suivantes : 

$$\psi(x,0)=\psi(x,1)=\psi(0,y)=\psi(1,y)=\frac{\partial\psi(x,0)}{\partial y}=\frac{\partial\psi(0,y)}{\partial x} = \frac{\partial\psi(1,y)}{\partial x}=0$$


$$\frac{\partial\psi(x,1)}{\partial y}=1$$

En dicrétisant les dérivées premières on trouves certaines égalités permettant de lever les inconnues aux bords. On introduit donc ces relations dans l'expression de A(p) :


```python
def A(psi):
    psic=psi.copy()
    psicc=psi.copy()
    
    psic[1,2:-2]=(1/4)*psicc[2,2:-2]
    psic[-2,2:-2]=(1/4)*psicc[-3,2:-2]
    psic[1:-1,-2]=(1/4)*psicc[1:-1,-3]
    psic[1:-1,1]=(1/4)*psicc[1:-1,2]
    
    psicc[2:-2,2:-2]=(psic[:-4,2:-2]+psic[2:-2,:-4]+psic[2:-2,4:]-8*psic[1:-3,2:-2]+20*psic[2:-2,2:-2]-8*psic[3:-1,2:-2]+psic[4:,2:-2]-8*psic[2:-2,3:-1]-8*psic[2:-2,1:-3]+2*psic[3:-1,1:-3]+2*psic[3:-1,3:-1]+2*psic[1:-3,1:-3]+2*psic[1:-3,3:-1])
    
    return psicc
```

```python
b0=numpy.zeros(shape=(nx,nx))
b0[2:-2,-3]=-numpy.ones(nx-4)*(dx*2)
b0[2:-2,-4]=numpy.ones(nx-4)*(dx/2)
psi=numpy.zeros(shape=(nx,nx))
```

Voici les valeurs

```python
nx = 41
dx = 1./(nx-1)
def L1norm(new, old):
    norm = numpy.sum(numpy.abs(new-old))
    return norm
```

```python
def poisson_2d_conjugate_gradient(p0, b,rtol=1e-8):   
    p = p0.copy()
    r = numpy.zeros_like(p)  # initial residual
    Ad = numpy.zeros_like(p)  # to store the mat-vec multiplication
    diff = rtol + 1  # initial difference
    # Compute the initial residual. here A(p)=0 and could be removed
    r = b - A(p)
    # Set the initial search direction to be the residual.
    d = r.copy()
    i=1
    while diff > rtol:
        pk = p.copy()
        rk = r.copy()
        # Compute the Laplacian of the search direction.
        Ad = A(d)
        alpha = numpy.sum(r * r) / numpy.sum(d * Ad)
        # Update the solution.
        p = pk + alpha * d
        # Update the residual.
        r = rk - alpha * Ad
        # Update the search direction.
        beta = numpy.sum(r * r) / numpy.sum(rk * rk)
        d = r + beta * d
        # Compute the relative L1-norm of the difference.
        diff = L1norm(p, pk)
        i+=1
    pkk=p.copy()  
    p[1:-1,-2]=(1/4)*pkk[1:-1,-3]-(dx/2)
    p[1,2:-2]=(1/4)*pkk[2,2:-2]
    p[1:-1,1]=(1/4)*pkk[1:-1,2]
    p[-2,2:-2]=(1/4)*pkk[-3,2:-2] 
    print('L1-norm=',diff) 
    print('#iteration=',i)
    return p
```

Ensuite il est demandé de trouver les valeurs de la fonction de flux pour les points (x,y) = (0.6,0.2), (0.6,0.4), (0.6,0.6), (0.6,0.8). Pour ce faire, il suffit de tirer les points voulu de notre liste de solutions. Etant donné que nous procédons par pas de 0.025, c.-à-d. la 8-ième, 16-ième, 24-ième et 32-ième itération respectivement pour y et la 24-ième itération pour x. Ceci nous donne :

```python
full_matrice=poisson_2d_conjugate_gradient(psi, b0, rtol=1e-8)
print('f(x,y)=f(0.6,0.2)=',full_matrice[24,8])
print('f(x,y)=f(0.6,0.4)=',full_matrice[24,16])
print('f(x,y)=f(0.6,0.6)=',full_matrice[24,24])    
print('f(x,y)=f(0.6,0.8)=',full_matrice[24,32])
```

```python
data = full_matrice
x = numpy.linspace(0, 1, nx)
y = numpy.linspace(0, 1, nx)
X,Y = numpy.meshgrid(x, y)
Z = data
fig, ax = pyplot.subplots()
p = ax.pcolor(Y, X, Z)
cb = fig.colorbar(p)
surf=pyplot.contour(Y,X,Z,10,cmap=cm.viridis)
fig.colorbar(surf)
```
